```python
import pandas as pd
```

```python
dataset=pd.read_csv("CKD.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
dataset['classification_yes'].value_counts()
```

```python
independent=dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
```

```python
dependent=dataset[['classification_yes']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
grid_param={'n_estimators':[100,500,1000,1500],'criterion':['gini', 'entropy', 'log_loss'],'max_features':['sqrt', 'log2', None]
            ,'bootstrap':[True, False]
            ,'warm_start':[True, False]}

RF_grid=GridSearchCV(RandomForestClassifier(), grid_param,refit=True, verbose=3,n_jobs=-1, scoring='f1_weighted')

RF_grid.fit(x_train,y_train)

RF_pred=RF_grid.predict(x_test)
```

```python
print("The best Hyperfactor Parameter are\n",RF_grid.best_params_)
re=RF_grid.cv_results_

from sklearn.metrics import confusion_matrix
cm_RF=confusion_matrix(y_test,RF_pred)

from sklearn.metrics import classification_report
class_report_RF=classification_report(y_test,RF_pred)
print("\nRandomForestClassifier:\n",class_report_RF)
```
